```python
#THÊM CỘT VÀO BẢNG ĐÃ CÓ - vị trí ở giữa 2 cột đã có
import pandas as pd
import numpy as np
data = pd.read_csv('FoodPrice_in_Turkey.csv', header = 0, encoding = 'utf-8')
data["tên cột mới"] = np.NaN
data.insert(7,"chiết khấu",20)
data.insert(9,"số lượng",50)
#đổi tên cột mới thành  "tổng" và có giá trị giá*số lượng*(100% - chiết khấu)

data2 = data.rename(columns = {'tên cột mới':'tổng'},inplace = True)
data["tổng"] = data["Price"]*data['số lượng']*(100-data['chiết khấu'])/100
#data ['tổng'] = data ['Giá'] * data ['Số lượng'] * (100-data ['Chiết khấu (%)'])/100

data
```
