```python
import pandas as pd
import numpy as np
data = pd.read_csv('OnlineRetail.csv', header = 0, encoding = "ISO-8859-1")
#in ra số hàng và số cột
print('Số hàng là:', data.shape[0], ', Số cột là:', data.shape[1])
#việt hóa tên cột
data.columns = ['Số hóa đơn', 'Mã hàng hóa', 'Mô tả', 'Số lượng', 'Ngày', 'Đơn giá', 'Mã KH', "Quốc gia"]
#thêm cột thành tiền với giá trị bằng 0
data['Thành tiền'] = 0
#thêm cột "chiết khấu" sau cột "đơn giá"
data.insert(6,'Chiết khấu (%)', 3)
#thêm 1 sp sau hàng có index 541905
#thêm 324345, 34534B, Noi chien khong dau, 1000, 23/6/2022 20:30, 50.2, 3, 312313.0, Vietnam, NaN
data.loc[541905.5] = [324345, "34534B", 'Noi chien khong dau', 1000, '23/6/2022 20:30', 50.2, 3, 312313.0, 'Vietnam', np.NaN]
data = data.sort_index()
#tính thành tiền = số lượng * đơn giá  chiết khấu
data.loc[:,"Thành tiền"] = data.loc[:,"Số lượng"]*data.loc[:,"Đơn giá"]*(100 - data.loc[:,"Chiết khấu (%)"])/100
#data.to_csv()
data
```
