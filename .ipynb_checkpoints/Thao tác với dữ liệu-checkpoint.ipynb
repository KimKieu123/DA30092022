```python
#Sinh D·ªØ Li·ªáu T·ª´ C√°c C·ªôt ƒê√£ C√≥
#T·ª™ PH√âP TO√ÅN S·ªê H·ªåC + - * /...
import pandas as pd
import numpy as np
data = pd.read_csv('shopeep_koreantop_clothing_shop_data.csv', header=0)
#th√™m c·ªôt rating = rang ting good * 2 + rating normal - rating bad *3
data.loc[:,"rating"] = data.loc[:,"rating_good"]*2 + data.loc[:,"rating_normal"] - data.loc[:,"rating_bad"]*3
data.loc[:,["rating_good","rating_normal","rating_bad"]]
```

```python
#th√™m c·ªôt location v·ªõi c·∫•u tr√∫c: shop_location, Philippine
data.loc[:,"location"] = data.loc[:,"shop_location"] + ", Philippine"
data
#n·∫øu d·ªØ li·ªáu NaN k·∫øt qu·∫£ v·∫´n l√† NaN
```

```python
#th√™m c·ªôt "month day, year" v·ªõi "jont_month" + "joint_day" + "joint_year"
#g·ª£i √Ω ƒë·ªïi th√°nh str v√¨ c√≥ th·ªÉ c√≥ d·ªØ li·ªáu l√† s·ªë
data.loc[:,'join'] = data.loc[:,'join_month'] + " " + data.loc[:,'join_day'].astype(str) + " " +data.loc[:,'join_year'].astype(str)
data.loc[:,'join']
```

```python
#C√ÅC PH√âP TO√ÅN LOGIC
#th√™m c·ªôt test1 v·ªõi c√°c shop m·ªü sau nƒÉm 2020
data.loc[:,"test1"] = data.loc[:,"joint_year"] > 2020
data.loc[:,["name",'join_year',"test1"]]
```
