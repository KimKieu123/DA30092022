```python
#Sinh Dữ Liệu Từ Các Cột Đã Có
#TỪ PHÉP TOÁN SỐ HỌC + - * /...
import pandas as pd
import numpy as np
data = pd.read_csv('shopeep_koreantop_clothing_shop_data.csv', header=0)
#thêm cột rating = rang ting good * 2 + rating normal - rating bad *3
data.loc[:,"rating"] = data.loc[:,"rating_good"]*2 + data.loc[:,"rating_normal"] - data.loc[:,"rating_bad"]*3
data.loc[:,["rating_good","rating_normal","rating_bad","rating"]]
```

```python
#thêm cột location với cấu trúc: shop_location, Philippine
data.loc[:,"location"] = data.loc[:,"shop_location"] + ", Philippine"
data
#nếu dữ liệu NaN kết quả vẫn là NaN
```

```python
#thêm cột "month day, year" với "jont_month" + "joint_day" + "joint_year"
#gợi ý đổi thánh str vì có thể có dữ liệu là số
data.loc[:,'join'] = data.loc[:,'join_month'] + " " + data.loc[:,'join_day'].astype(str) + " " +data.loc[:,'join_year'].astype(str)
data.loc[:,'join']
```

```python
#CÁC PHÉP TOÁN LOGIC
#thêm cột test1 với các shop mở sau năm 2020
data.loc[:,"test1"] = data.loc[:,"join_year"] > 2020
data.loc[:,["name",'join_year',"test1"]]
```

```python
#GÁN GIÁ TRỊ THEO KIỂU IF THEN ELSE
#Thêm cột "medal" với giá trị "gold" nếu rating > 10000 và "silver"nếu <=10000
data.loc[:,'medal'] = np.where(data.loc[:,'rating'] > 10000, 'Gold', 'Silver')
display(data.loc[:,['name','rating','medal']])
```

```python
#GÁN GIÁ TRỊ THEO KIỂU SWICH ... CASE
#Sửa cột 'medal' với giá trị là 'gold' nếu có điểm rating =>100000 và 'silver' nếu 50000<=rating<100000 và 'bronze' nếu 10000<=rating<50000 và NaN nếu rating <10000
conditions = [(data.loc[:,'rating']>=100000),
              (data.loc[:,'rating']>=50000) & (data.loc[:,'rating']<100000),
              (data.loc[:,'rating']>=10000) & (data.loc[:,'rating']<50000)]
values = ['gold','silver','bronze']
data.loc[:,'medal']=np.select(conditions, values, default=np.NaN)
data.loc[:,['name','rating','medal']]
```

```python
#BUỔI 3 NGÀY 19/10/2022
#TÁCH CỘT THÀNH CÁC CỘT NHỎ HƠN
import pandas as pd
import numpy as np
data = pd.read_csv('shopeep_koreantop_clothing_shop_data.csv', header=0)
data.loc[:,"shop_location"].str.split(",")
#là series bao gồm list
```

```python
temp = data.loc[:,"shop_location"].str.split(",")
data.loc[:,"distric"] = temp.str[0]
data.loc[:,"city"] = temp.str[1]
data.loc[:,["shop_location","distric","city"]]
```

```python
#đối với dữ liệu là thời gian
data.loc[:,"date_collected"]
#đã theo format ngày tháng nhưng dtype đang ở dạng str
```

```python
pd.to_datetime(data.loc[:,"date_collected"], format = "%Y-%m-%d")
```

```python
#pd.to_datetime(data.loc[:,"date_collected"], format = "%Y-%m-%d").dt.day
#pd.to_datetime(data.loc[:,"date_collected"], format = "%Y-%m-%d").dt.month
#pd.to_datetime(data.loc[:,"date_collected"], format = "%Y-%m-%d").dt.year
temp2 = pd.to_datetime(data.loc[:,"date_collected"], format = "%Y-%m-%d")
data.loc[:,"day"] = temp2.dt.day
data.loc[:,"month"] = temp2.dt.month
data.loc[:,"year"] = temp2.dt.year
data.loc[:,["day","month","year"]]
```

```python
data.loc[:,"response_time"].head()
temp3 = pd.to_datetime(data.loc[:,"response_time"], format = " %H:%M:%S") #dữ liệu gốc có dấu cách phí trước nên format cũng phải có
temp3.head()
data.loc[:,"Hour"] = temp3.dt.hour
data.loc[:,"Minute"] = temp3.dt.minute
data.loc[:,"Second"] = temp3.dt.second
data.loc[:,["Hour","Minute","Second"]]
```
