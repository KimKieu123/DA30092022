```python
#kêt hợp nhiều biểu đồ cùng loại

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#Tạo dữ liệu nhân tạo
y1 = np.random.normal(200, 30, size = 100) 
#100 giá trị dao động xung quanh 200 +- 30
y2 = np.random.normal(100, 70, size = 100)
data = pd.DataFrame({'y1': y1, 'y2': y2, 'date': pd.date_range('1/1/2021', periods=100)})
data
```

```python
#cũng loại trên 1 Axes
plt.plot(data.index, data.loc[:,"y1"], "r", label = "y1") #màu đỏ
plt.plot(data.index, data.loc[:,"y2"], "b", label = "y2") #màu xanh

plt.title("đồ thị phân bổ normail")
plt.xlabel("index")
plt.ylabel("dao động")
plt.legend() #hiện theo label
plt.show
```

```python
#Tạo dữ liệu nhân tạo
y3 = np.random.normal(200, 30, size = 100)
y4 = np.random.normal(1000, 200, size = 100)
data = pd.DataFrame({'y3': y3, 'y4': y4, 'date': pd.date_range('1/1/2021', periods=100)})
data
#Vẽ đồ thị ghép (cùng loại trên cùng 1 axes) nên tránh
plt.plot(data.index, data.loc[:,'y3'], 'r', label='Giá xe') #Đường y3 màu đỏ
plt.plot(data.index, data.loc[:,'y4'], 'b', label='Doanh số') #Đường y4 màu xanh

plt.title('Đồ thị phân bố normal') #Tên đồ thị
plt.xlabel('Index') #Tên trục X
plt.ylabel('Dao động') #Tên trục Y
plt.legend() #Legend (chỉ nên dùng khi có từ 2 đồ thị con trở lên)
plt.show()

#không cùng mô tả về 1 loại dữ liệu
#chênh lệch giữa 2 nhóm giá trị quá lớn
```

```python
#nhiều biểu đồ cùng loại nhiều axes trên 1 figure
#Tạo dữ liệu
x1 = ['Day 1','Day 2','Day 3','Day 4','Day 5','Day 6','Day 7','Day 8','Day 9','Day 10']
y1 = [12.2, 16.3, 20.5, 25.4, 31.2, 24.2, 18, 13.5, 17.1, 21.7]
y2 = [20.3, 27.6, 31.4, 33.7, 39.2, 33.4, 29.2, 25.7, 31.2, 36]
y3 = [16.3, 17.6, 24.4, 26.7, 33.2, 29.4, 27.2, 24.7, 27.2, 32]
data2 = pd.DataFrame({"x":x1,"y1":y1,"y2":y2,"y3":y3})
data2
```

```python
#tạo figure và subplot (các axes = ax)
fig1, ax = plt.subplots(3,1) #3 hàng 1 cột
```

```python
fig1, ax = plt.subplots(3,1)
ax[0].plot(data2["x"], data2["y1"], label = "y1")
ax[1].plot(data2["x"], data2["y2"], label = "y2")
ax[2].plot(data2["x"], data2["y3"], label = "y3")

#đặt title trên cùng cho bảng 1
ax[0].set_title("giá các mặt hàng") 
ax[2].set_xlabel("ngày")

ax[0].set_ylabel("giá vàng")
ax[1].set_ylabel("giá bạc")
ax[2].set_ylabel("giá đồng")

#xóa tên ngày cho khung 1 và 0, chỉ hiện khung dưới cùng
ax[0].set_xticklabels([])
ax[1].set_xticklabels([])

ax[0].tick_params(axis='y', size = 8)
ax[1].tick_params(axis='y', size = 8)
ax[2].tick_params(axis='y', size = 8)

ax[0].tick_params(axis='x', size = 2)
ax[1].tick_params(axis='x', size = 2)
ax[2].tick_params(axis='x', size = 8)

#ẳn các stick cho 0 và 1
# ax[0].set_xticks([]) #Remove tick
# ax[1].set_xticks([]) #Remove tick

#thay đổi khoảng cách giữa các khung
plt.tight_layout(pad = 1)

plt.show()
```

```python
fig1, ax = plt.subplots(3,1, sharex = True, sharey = True)
#sharex : vd: 1 dữ liệu 11 tháng, 1 dữ liệu 12 tháng ==> trình bày bằng nhau
#sharey: khi dữ liệu scale gần bằng nhau
#thay có việc set các tick_params
ax[0].plot(data2["x"], data2["y1"], label = "y1")
ax[1].plot(data2["x"], data2["y2"], label = "y2")
ax[2].plot(data2["x"], data2["y3"], label = "y3")

#đặt title trên cùng cho bảng 1
ax[0].set_title("giá các mặt hàng") 
ax[2].set_xlabel("ngày")

ax[0].set_ylabel("giá vàng")
ax[1].set_ylabel("giá bạc")
ax[2].set_ylabel("giá đồng")

#xóa tên ngày cho khung 1 và 0, chỉ hiện khung dưới cùng
ax[0].set_xticklabels([])
ax[1].set_xticklabels([])

#ẳn các stick cho 0 và 1
# ax[0].set_xticks([]) #Remove tick
# ax[1].set_xticks([]) #Remove tick

#thay đổi khoảng cách giữa các khung
plt.tight_layout(pad = 1)

plt.show()
```

```python
#Tạo dữ liệu nhân tạo
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
y1 = np.random.normal(100, 30, size = 100)
y2 = np.random.normal(100, 70, size = 100)
y3 = np.random.normal(200, 30, size = 100)
y4 = np.random.normal(1000, 200, size = 100)
data3 = pd.DataFrame({'y1':y1, 'y2':y2, 'y3': y3, 'y4': y4})
data3
#vẽ đồ thị 2*2 
#trục x xuất phát từ 0, cách nhau 10 giá trị, trục x là indexx
#tên là "VD3"
#tên x là "index"
#ten y là y1,y2...
#màu: r, g, b, y
fig1, ax = plt.subplots(2,2, sharex = True)
ax[0][0].plot(data3.index, data3["y1"],"r", label = "y1")
ax[0][1].plot(data3.index, data3["y2"],"g", label = "y2")
ax[1][0].plot(data3.index, data3["y3"],"b", label = "y3")
ax[1][1].plot(data3.index, data3["y4"],"y", label = "y4")

ax[0][0].set_ylabel('y1')
ax[0][1].set_ylabel('y2')
ax[1][0].set_ylabel('y3')
ax[1][1].set_ylabel('y4')

ax[1][0].set_xlabel('My Index')
ax[1][1].set_xlabel('My Index')

#Đặt tên cho từng axes
# ax[0][0].set_title('Đây là màu đỏ')
# ax[0][1].set_title('Đây là màu lam')
# ax[1][0].set_title('Đây là màu lục')
# ax[1][1].set_title('Đây là màu vàng')

#Thêm ledend cho mỗi axes
ax[0][0].legend()
ax[0][1].legend()
ax[1][0].legend()
ax[1][1].legend()

#Đặt tên cho cả figure
plt.suptitle('Ví dụ 3')
ax = plt.gca()
ax.set_xlim(left = 0, right = 100)
ax.xaxis.set_major_locator(MultipleLocator(10))

plt.tight_layout() #Tự động căn chỉnh đồ thị cho đẹp


plt.show()
```
