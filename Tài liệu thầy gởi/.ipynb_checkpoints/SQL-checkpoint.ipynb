```python
#Truy xuáº¥t tÃªn khÃ¡ch hÃ ng, nÆ°á»›c vÃ  thÃ nh phá»‘
# mÃ  nÆ°á»›c lÃ  Germany hoáº·c France
query = "SELECT customerName, country, city FROM customers \
WHERE country = 'Germany' OR country = 'France'"
vidu5 = pd.read_sql_query(query, connection)
vidu5
```

```python
#Truy xuáº¥t tÃªn khÃ¡ch hÃ ng, nÆ°á»›c vÃ  thÃ nh phá»‘
# mÃ  nÆ°á»›c khÃ´ng pháº£i lÃ  Germany
query = "SELECT customerName, country, city FROM customers \
WHERE NOT country = 'Germany'"
vidu6 = pd.read_sql_query(query, connection)
vidu6
```

```python
#Truy xuáº¥t tÃªn khÃ¡ch hÃ ng, nÆ°á»›c vÃ  thÃ nh phá»‘
# mÃ  nÆ°á»›c lÃ  Germany, vÃ  thÃ nh phá»‘ lÃ  Berlin hoáº·c Munich
query = "SELECT customerName, country, city FROM customers \
WHERE country = 'Germany' AND (city = 'Berlin' OR city = 'Munich')      "
vidu6 = pd.read_sql_query(query, connection)
vidu6
```

```python
#Truy xuáº¥t tÃªn khÃ¡ch hÃ ng, nÆ°á»›c vÃ  thÃ nh phá»‘
# mÃ  nÆ°á»›c lÃ  Germany
# sáº¯p xáº¿p theo thá»© tá»± ABC tÃªn khÃ¡ch hÃ ng
query = "SELECT customerName, country, city FROM customers \
WHERE country = 'Germany' \
ORDER BY customerName ASC"
vidu8 = pd.read_sql_query(query, connection)
vidu8
```

```python
#Truy xuáº¥t tÃªn khÃ¡ch hÃ ng, nÆ°á»›c vÃ  thÃ nh phá»‘
# mÃ  nÆ°á»›c lÃ  Germany hoáº·c USA
# sáº¯p xáº¿p theo thá»© tá»± ABC tÃªn nÆ°á»›c
# sáº¯p xáº¿p theo thá»© tá»± ABC tÃªn thÃ nh phá»‘
query = "SELECT customerName, country, city FROM customers \
WHERE country = 'Germany' OR country = 'USA' \
ORDER BY country, city ASC"
vidu8 = pd.read_sql_query(query, connection)
vidu8
```
