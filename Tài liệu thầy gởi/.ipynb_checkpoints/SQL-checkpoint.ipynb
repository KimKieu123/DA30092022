```python
#Truy xuất tên khách hàng và thành phố mà thành phố có 3 chữ cái và bắt đầu bằng chữ N
query = "SELECT customerName, city FROM customers WHERE city LIKE 'N__'" #Dấu % k dùng được trong sqlalchemy
vidu3 = pd.read_sql_query(query, connection)
vidu3
```

```python
#Truy xuất tên khách hàng, nước và thành phố
# mà nước là Germany và city là Berlin
query = "SELECT customerName, country, city FROM customers \
WHERE country = 'Germany' AND city = 'Berlin'"
vidu4 = pd.read_sql_query(query, connection)
vidu4
```

```python
#Truy xuất tên khách hàng, nước và thành phố
# mà nước là Germany hoặc France
query = "SELECT customerName, country, city FROM customers \
WHERE country = 'Germany' OR country = 'France'"
vidu5 = pd.read_sql_query(query, connection)
vidu5
```

```python
#Truy xuất tên khách hàng, nước và thành phố
# mà nước không phải là Germany
query = "SELECT customerName, country, city FROM customers \
WHERE NOT country = 'Germany'"
vidu6 = pd.read_sql_query(query, connection)
vidu6
```

```python
#Truy xuất tên khách hàng, nước và thành phố
# mà nước là Germany, và thành phố là Berlin hoặc Munich
query = "SELECT customerName, country, city FROM customers \
WHERE country = 'Germany' AND (city = 'Berlin' OR city = 'Munich')      "
vidu6 = pd.read_sql_query(query, connection)
vidu6
```

```python
#Truy xuất tên khách hàng, nước và thành phố
# mà nước là Germany
# sắp xếp theo thứ tự ABC tên khách hàng
query = "SELECT customerName, country, city FROM customers \
WHERE country = 'Germany' \
ORDER BY customerName ASC"
vidu8 = pd.read_sql_query(query, connection)
vidu8
```

```python
#Truy xuất tên khách hàng, nước và thành phố
# mà nước là Germany hoặc USA
# sắp xếp theo thứ tự ABC tên nước
# sắp xếp theo thứ tự ABC tên thành phố
query = "SELECT customerName, country, city FROM customers \
WHERE country = 'Germany' OR country = 'USA' \
ORDER BY country, city ASC"
vidu8 = pd.read_sql_query(query, connection)
vidu8
```
