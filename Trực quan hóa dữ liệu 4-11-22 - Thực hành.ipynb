```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('GDPlist.csv', encoding = "ISO-8859-1")
data.head()
#so sánh GDP các nước 'South America
SA = data.loc[data.loc[:,'Continent'] == 'South America']
#sort lại giá trị 
SA = SA.sort_values(by = ["GDP (millions of US$)"])
plt.bar(x = SA.loc[:,"Country"], height = SA.loc[:,"GDP (millions of US$)"]/1000)
plt.xticks(rotation = 45)
plt.title("So sánh GPD các quốc gia Nam Mỹ")
plt.ylabel("tỷ USD")
plt.show()
#trong 9 nước Nam Mỹ thì Brazil là nước có tổng Sp quốc nội cao nhất 
#vượt lên hẳn các nước còn lại
# ngoài ra các nước còn lại có thể đc chia thành 2 nhóm
#nhóm trung bình : Peru, Chi le, Venezuela, Colombia, Argentina
#nhóm thấp: Paraguay, Uruguay, Ecudor
```

```python
#đánh giá tỷ lệ đóng góp của VN trên tổng : VN,Indo, Cam, Thai, Malay
sea_ = ["\xa0Vietnam","\xa0Indonesia","\xa0Cambodia","\xa0Thailand", "\xa0Malaysia"]
SEA = data.loc[data.loc[:,"Country"].isin(sea_)]
plt.pie(SEA.loc[:,"GDP (millions of US$)"], labels = SEA.loc[:,"Country"],
       autopct = "%1.2f%%")
plt.title("So sánh GPD các nước Đông Nam Á")
plt.show()
```
