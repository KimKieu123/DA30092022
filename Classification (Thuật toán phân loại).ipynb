```python
#Demo 3
#phân loại nhiều nhóm, mỗi đối tượng có thể nằm trong 1 hoặc nhiều nhãn
#VD phân loại phim: 1 bộ phim có thể thuộc nhiều thể loại khác nhau (kinh dị, trinh thám...)

# example of multi-class classification task
from numpy import where
from collections import Counter
from sklearn.datasets import make_blobs
from matplotlib import pyplot
# define dataset
X, y = make_blobs(n_samples=600, centers=2, random_state=3, cluster_std = 3.0)
# summarize dataset shape
print(X.shape, y.shape)
# summarize observations by class label
counter = Counter(y)
print(counter)
# summarize first few examples
for i in range(10):
	print(X[i], y[i])
# plot the dataset and color the by class label
for label, _ in counter.items():
	row_ix = where(y == label)[0]
	pyplot.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label))
pyplot.legend()
pyplot.show()
```

```python
import warnings
warnings.filterwarnings("ignore") #ko hiển thị các cảnh báo
# from numpy import where
# from collections import Counter
from sklearn.datasets import make_blobs
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
#đánh giá regression thay vì r2score
from sklearn.metrics import classification_report, confusion_matrix

# Tạo dataset
X, Y = make_blobs(n_samples=600, centers=2, random_state=3, cluster_std=2)

# sns.scatterplot(X[:,0], X[:,1])
# plt.show()

sns.scatterplot(x=X[:,0], y=X[:,1], hue=Y)
plt.show()

```

```python
import sklearn.model_selection as model_selection
#Chia bộ dữ liệu thành 2 phần train và test
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X,Y, train_size=0.6)
# X_train
```

```python
# instantiate the model (using the default parameters)
# logreg = LogisticRegression(random_state=16)
logreg = LogisticRegression()

# fit the model with data
logreg.fit(X_train, Y_train)

#Dự đoán với giá trị đầu vào là X_test
Y_pred = logreg.predict(X_test)

sns.scatterplot(x=X_train[:,0], y=X_train[:,1], hue=Y_train)
plt.show()

sns.scatterplot(x=X_test[:,0], y=X_test[:,1], hue=Y_test)
plt.show()

sns.scatterplot(x=X_test[:,0], y=X_test[:,1], hue=Y_pred)
plt.show()
```

```python
#Predict giá trị
import numpy as np
x_ = [5, -2]
#đưa về cùng dạng ma trận với x_test
x_ = np.array(x_).reshape(1, -1)
# x_
y_ = logreg.predict(x_)
print(y_)
# Biểu diễn điểm này trên đồ thị
sns.scatterplot(x=X_test[:,0], y=X_test[:,1], hue=Y_test)
sns.scatterplot(x=x_[:,0], y=x_[:,1], color='r')
plt.show()
```

```python
#Evaluate model
#trả về 1 ma trận nhị phân
conf_m = confusion_matrix(Y_test, Y_pred)
#bảng ngẫu nhiên
report = classification_report(Y_test, Y_pred)

print('conf_m:', conf_m, sep='\n', end='\n\n')
sns.heatmap(conf_m)
plt.show()

print('report:', report, sep='\n')
```

```python
#Thư viện
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
from sklearn.model_selection import train_test_split
import time
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
dat2 = pd.read_csv("mobile train.csv")
X = dat2.drop(columns='price_range',axis=1)
Y = dat2['price_range']
X_train,X_test,Y_train,Y_test = train_test_split(X, Y, test_size=0.2)


#Tạo model
dec_tree = DecisionTreeClassifier()

#fit, train
start = time.time()
dec_tree.fit(X_train,Y_train)
print('Đã học xong trong:', time.time()-start, ' s')

#dự đoán

Y_pred = dec_tree.predict(X_test)


#Tạo bảng df để trực quan hóa
test = pd.concat([X_test, Y_test], axis=1)
Y_pred = pd.DataFrame(Y_pred, index=test.index, columns = ['price_range'])
pred = pd.concat([X_test, Y_pred], axis=1)

#Evaluate model
conf_m = confusion_matrix(Y_test, Y_pred)
report = classification_report(Y_test, Y_pred)

print('conf_m:', conf_m, sep='\n', end='\n\n')
sns.heatmap(conf_m)
plt.show()

print('report:', report, sep='\n')

```
