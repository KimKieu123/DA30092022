```python
#[Thực hành] Thêm cột dữ liệu mới từ các cột dữ liệu đã có
import pandas as pd
df=pd.read_csv('shopeep_koreantop_clothing_shop_data.csv')
# Để đơn giản ta thực hiện lọc tập dữ liệu ban đầu theo các thuộc tính sau
df = df[['join_month', 'join_day','join_year','shop_location','rating_bad','rating_good','rating_normal']]
# Tạo cột 'rating' tính điểm cho mỗi cửa hàng dựa vào thông tin các cột rating_bad, rating_good, rating_normal 
# theo công thức sau: rating = rating_good *2 + rating_normal - rating_bad*3

df['rating'] = df['rating_good'] * 2 + df['rating_normal'] - df['rating_bad'] * 3
# ghép 3 cột join_month, join_day, join_year thành cột mới có tên 'date' nhận giá trị có dạng: "join_month join_day,join_year"

df['date'] = df['join_month'] + " " + df['join_day'].astype(str) + "," + df['join_year'].astype(str)
# Thêm cột new có giá trị True nếu join_year = 2021 và False trong trường hợp còn lại.
df['new'] = df['join_year'] == 2021 
# Nếu có 2 lựa chọn ta sử dụng np.where
# Thêm cột rate có giá trị good nếu rating_good >= 50000,  bad trong trường hợp còn lại

import numpy as np
df['rate'] = np.where(df['rating_good'] >= 50000,'good','bad' )
# Nếu có nhiều hơn 2 lựa chọn ta sử dụng np.select
# Thêm cột flag tặng cờ cho các cửa hàng, flag nhận các giá trị như sau:
# blue khi rating_good >= 30000 và rating_bad <= 100
# yellow khi 10000 <= rating_good < 30000 và 100 < rating_bad <= 1000
# red khi rating_good < 10000
# black đối với các trường hợp còn lại

conditions = [(df['rating_good'] >= 30000) & (df['rating_bad'] <= 100),
              (df['rating_good'] >= 10000) & (df['rating_good'] < 30000) & (df['rating_bad'] <= 1000) & (df['rating_bad'] > 100),
              (df['rating_good'] < 10000)]
choices = ['blue', 'yellow','red']
df['flag'] = np.select(conditions, choices, default='black')
df 
```

```python
#[Bài tập] Phân tích dữ liệu Online Retail
#Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.
import pandas as pd
import numpy as np
from datetime import datetime
df=pd.read_csv('OnlineRetail.csv', encoding = "ISO-8859-1")
print (df.info())
#Tạo cột mới có tên quý –  ‘Previous’ nhận giá trị 1 nếu ngày lập hóa đơn nằm trong các tháng 1,2,3; 
#nhận giá trị 2 nếu ngày lập hóa đơn nằm trong các tháng 4,5,6; 
#nhận giá trị 3 nếu ngày lập hóa đơn nằm trong các tháng 7,8,9;  
#nhận giá trị 4 nếu ngày lập hóa đơn nằm trong các tháng 10,11,12;
conditions = [(pd.to_datetime(df['InvoiceDate'],format='%m/%d/%Y %H:%M').dt.month <= datetime.strptime("3", "%m").month),
             (pd.to_datetime(df['InvoiceDate'],format='%m/%d/%Y %H:%M').dt.month <= datetime.strptime("6", "%m").month),
             (pd.to_datetime(df['InvoiceDate'],format='%m/%d/%Y %H:%M').dt.month <= datetime.strptime("9", "%m").month)]  
choices = [1,2,3]
df['Previous'] = np.select(conditions, choices, default= 4)
#Tạo một cột mới có tên ‘Amount’ có giá trị bằng Quantity * UnitPrice
df["Amount"] = df["Quantity"]*df["UnitPrice"]
#Tạo cột mới ‘Discount’ nhận giá trị 10% nếu Country là ‘United Kingdom’ và thuộc quý 4, 5% nếu
#là ‘France’ ngược lại là 0%.

conditions = [(df["Country"].astype(str) == "United Kingdom") & (df["Previous"] == 4),
              (df["Country"].astype(str) == "France") & (df["Previous"] == 4)]
choices = [0.1, 0.05]
df["Discount"] = np.select(conditions,choices, default = 0)

#Tạo cột mới ‘Total’ nhận giá trị bằng: Amount – Discount.


df["Total"] = df["Amount"] - df["Amount"]*df["Discount"]

df
```

```python
#[Bài tập] Phân tích bộ dữ liệu HousePrice _ Đống Đa
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_excel('house_price_dống-da.xlsx')
#Vẽ biểu đồ phân tích mối liên hệ giữa diện tích với giá nhà, giữa số phòng ngủ với giá nhà,
#giữa số toilet với giá nhà.
fig1, ax = plt.subplots(3,1)

ax[0].plot(data["price"], data["area"], label = "Price-area")
ax[1].plot(data["price"], data["bedroom"], label = "Price-bedroom")
ax[2].plot(data["price"], data["toilet"], label = "Price-toilet")

ax[0].set_title("mối liên hệ giữa giá nhà") 
ax[2].set_xlabel("price")

ax[0].set_ylabel("area")
ax[1].set_ylabel("bedroom")
ax[2].set_ylabel("toilet")

ax[0].set_xticklabels([])
ax[1].set_xticklabels([])

plt.tight_layout(pad = 1)
plt.show()

#Vẽ biểu đồ so sánh giá nhà trung bình trên 1 m2 giữa các hình thức nhà (type_of_land).
data['type_of_land'].replace("Bán nhà mặt phố\n ","Bán nhà mặt phố",inplace = True)
data['type_of_land'].replace("Bán nhà riêng\n ","Bán nhà riêng",inplace = True)
data['type_of_land'].replace("Bất động sản khác\n ","Bất động sản khác",inplace = True)
#data['type_of_land'].unique()
data.drop(data.loc[data["area"] == 0].index,axis = 0,inplace = True)
data.dropna(subset=['price'],inplace=True)
data.dropna(subset=['area'],inplace=True)
data["giá trung bình"] =  data['price']/data["area"]
d1 = data.groupby('type_of_land')['giá trung bình'].mean()
d1 = pd.DataFrame(d1)
plt.bar(x = d1.index, height = d1.loc[:,"giá trung bình"],width = 0.5)
plt.xticks(rotation = 45)
plt.title("So sánh giá và hình thức nhà")
plt.ylabel("giá trung bình")
plt.show()

#Vẽ biểu đồ thể hiện tỉ lệ % bài đăng (bản ghi) giữa các hình thức nhà (type_of_land).
d2 = data.groupby('type_of_land')['postDate'].count()
d2 = pd.DataFrame(d2)
plt.pie(d2.loc[:,'postDate'], labels = d2.index,
       autopct = "%1.2f%%")
plt.title("So sánh % bài đăng")
plt.show()

#Vẽ biểu đồ thể hiện sự thay đổi giá nhà trung bình trên 1m2 theo số lượng phòng ngủ.
data.dropna(subset=['bedroom'],inplace=True)
d3 = data.groupby('bedroom')['giá trung bình'].mean()
d3 = pd.DataFrame(d3)
x = d3.index.get_level_values(0)
plt.plot(x, d3['giá trung bình'])
plt.title('Thay đổi giá so với số lượng phòng ngủ', fontsize = 16)
plt.xlabel('số lượng phòng ngủ', fontsize = 14)
plt.ylabel('giá trung bình', fontsize = 14)
plt.show()
```

```python
#[Bài tập] Phân tích bộ dữ liệu Shopee.ph
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('shopeep_koreantop_clothing_shop_data.csv')

#Vẽ biểu đồ so sánh số lượng shop gia nhập theo các năm.
d1 = df.groupby('join_year')['name'].count()
d1 = pd.DataFrame(d1)
plt.bar(x = d1.index, height = d1.loc[:,"name"],width = 0.5)
plt.xticks(rotation = 45)
plt.title("so sánh số lượng shop gia nhập theo các năm")
plt.ylabel("số lượng shop")
plt.show()

#Vẽ biểu đồ thể hiện mối quan hệ giữa tỉ lệ phản hồi với số lượt khách hàng đánh giá tốt
d2 = df.groupby('response_rate')['rating_good'].sum()
d2 = pd.DataFrame(d2)
x = d2.index.get_level_values(0)
plt.plot(x, d2['rating_good'])
plt.title('mối quan hệ giữa tỉ lệ phản hồi với số lượt khách hàng đánh giá tốt', fontsize = 16)
plt.xlabel('response_rate', fontsize = 14)
plt.ylabel('rating_good', fontsize = 14)
plt.show()

#Vẽ biểu đồ thể hiện mối quan hệ giữa thời gian phản hồi (đơn vị giây) với số lượt khách hàng đánh giá xấu.
df['Hour']=pd.to_datetime(df['response_time'],format=' %H:%M:%S').dt.hour
df['Minute']=pd.to_datetime(df['response_time'],format=' %H:%M:%S').dt.minute
df['Second']=pd.to_datetime(df['response_time'],format=' %H:%M:%S').dt.second
df["thời gian phản hồi"] = df['Hour']*60*60 + df['Minute']*60 + df['Second']
d3 = df.groupby('thời gian phản hồi')['rating_bad'].sum()
d3 = pd.DataFrame(d3)
colors = np.random.randint(0, 100, d3['rating_bad'].count())
sizes = np.random.randint(3, 7, d3['rating_bad'].count())
plt.scatter(d3.index, d3['rating_bad'],c=colors, s=sizes)
plt.title('Relationship between response_time and rating_bad', fontsize = 16, color = 'r')
plt.xlabel('response_time', fontsize = 14)
plt.ylabel('rating_bad', fontsize = 14)
plt.show()

#Vẽ biểu đồ thể hiện xu hướng của số lượng shop gia nhập theo thời gian.
x = d1.index.get_level_values(0)
plt.plot(x, d1['name'])
plt.title('xu hướng của số lượng shop gia nhập theo thời gian', fontsize = 16)
plt.xlabel('năm', fontsize = 14)
plt.ylabel('số lượng shop', fontsize = 14)
plt.show()

#Vẽ biểu đồ thể hiện phân bố của điểm đánh giá trung bình. 
plt.hist(df["rating_star"], bins = 50)
plt.show()
```
