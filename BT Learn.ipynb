```python
#[Thực hành] Thực hành trên bộ dữ liệu covid 19
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('subset-covid-data.csv', encoding= 'UTF-8')
data.head()
data.info()
# Tìm hiểu xem dữ liệu được thống kê cho những ngày nào
data.date.value_counts()
# lọc dữ liệu nhiễu:
cleaned_data = data[data.date == '2020-04-12']
# Vẽ biểu đồ phân bố số lượng ca mắc mới ở các quốc gia
print ("trung bình số ca mắc mới: " + str(cleaned_data.cases.mean()))
print ("trung vị của số ca mắc mới: "+ str(cleaned_data.cases.median()))

plt.hist(cleaned_data.cases, bins = 200)
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title("Phân bố số ca mắc mới",fontdict = font1)
plt.xlabel("số số ca mắc mới",fontdict = font2)
plt.ylabel("Số lượng quốc gia",fontdict = font2)
print("tổng số ca nhiễm và số ca ncủa các châu lục")
cleaned_data.groupby('continent')['cases','deaths'].sum()
print ("5 quốc gia có số ca nhiễm mới cao nhất")
data = data.sort_values('cases',ascending = False)
data.head(5)
print ("5 quốc gia có số ca tử vong cao nhất")
data = data.sort_values('deaths',ascending = False)
data.head(5)
#Text(0, 0.5, 'Số lượng quốc gia')
```

```python
#[Bài tập] Phân tích trên bộ dữ liệu GPD List
import pandas as pd
data=pd.read_csv('GDPlist.csv', encoding = "ISO-8859-1")
#Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.
data.info()
#Việt hóa tên các cột trong bảng dữ liệu: Country 🡪 Nuoc; Continent 🡪 Chauluc; GDP (millions of US$) 🡪 GDP (trieu $)
data.rename(columns = {"Country":"Nuoc","Continent":"Chauluc","GDP (millions of US$)":"GDP (trieu $)"}, inplace = True)
#Chèn thêm một cột “Thanhpho” vào sau cột “Nuoc”, giá trị ban đầu là giá trị của cột “Nuoc”  
data.loc[:,"Thanhpho"] = data.loc[:,"Nuoc"]
#Trong cột Thanhpho, thay giá trị Vietnam thành Hanoi; Làm tương tự với các nước còn lại.
#data.loc[:,"Thanhpho"].replace("Vietnam","Hanoi",inplace = True)
data.loc[:, 'Thanhpho'].replace("Vietnam", 'Hanoi', inplace = True)
#Xóa các bản ghi có Chauluc là ‘Asia’
data.drop(data.loc[data["Chauluc"] == "Asia"].index,axis = 0,inplace = True)
#Xóa các bản ghi có GDP < 300000
data.drop(data.loc[data["GDP (trieu $)"] < 300000].index,axis = 0,inplace = True)
data
```

```python
#[Bài tập] Phân tích trên bộ dữ liệu GPD List
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('GDPlist.csv', encoding = "ISO-8859-1")
#Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.
data.info()
#Tính giá trị lớn nhất và nhỏ nhất của GDP.
print("giá trị nhỏ nhất",data["GDP (millions of US$)"].min())
print("giá trị lớn nhất",data["GDP (millions of US$)"].max())
#Hãy cho biết xu hướng phân bố dữ liệu của GDP.
print("giá trị trung vị",data["GDP (millions of US$)"].median())
print("giá trị trung bình",data["GDP (millions of US$)"].mean())
print("giá trị mode",data["GDP (millions of US$)"].mode())
plt.hist(data["GDP (millions of US$)"], bins = 200)
#dữ liệu lệch phải
#Hãy cho biết châu lục nào xuất hiện nhiều nhất?
#print("châu lục xuất hiện nhiều nhất",data.pivot_table(values='Country',index='Continent',aggfunc ='count'))
data_continent = data.groupby(["Continent"])['Country'].count().sort_values(ascending= False)
print("châu lục xuất hiện nhiều nhất",data_continent.head(1))
#Với mỗi châu lục hãy tính tổng GDP; trung bình cộng GDP.
#Hợp nhất 2 bảng này thành một bảng duy nhất gồm 3 thông tin: Tên châu lục; Tổng GDP; TBC GDP.
data1 = data.pivot_table(values='GDP (millions of US$)',index='Continent',aggfunc ='sum')
data2 = data.pivot_table(values='GDP (millions of US$)',index='Continent',aggfunc ='mean')
data3 = pd.concat([data1,data2], axis=1)
data3.rename(columns={'GDP (millions of US$)':'sum','GDP (millions of US$)':'mean'},inplace=True)

data3
```
