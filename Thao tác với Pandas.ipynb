```python
 #df.quantile(x,axis,interpolation ==> kiểm tra tính liền mạch của dữ liệu
# ==> chia lại dữ liệu (top 25%, 50%...)
# ==> lọc ra những giá trị ngoại lai (bất thường)
# x là số phân vij 0 <= x <= 1 (0%,25%,50%,100%)
# interpolation = "linear", "lower", "higher", "midpoint","nearest"
# ==> nội suy : khi mà dữ liệu thật ko có giá trị nào trùng trong khoảng x
# linear: tính trung bình của giá trị min và max
# lower lấy giá trị nhỏ nhất
# higher lấy giá trị cao nhất
# midpoint là giá trị ở giữa 
# nearest là giá trị gần nhất
# mặc định linear
data.loc[:,'Price'].quantile(0.5, "linear") # 0.5 tương đương 50% bằng giá trị median
```
